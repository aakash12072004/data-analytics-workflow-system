```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import HTML
```

```python
data = pd.read_csv("student_scores_dataset.csv")
data.head()
html_table = data.to_html(index=False, justify='center')

# Step 6: Display in notebook
display(HTML(html_table))
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data["Average"] = (data["Math_Score"] + data["Science_Score"] + data["English_Score"]) / 3
```

```python
plt.hist(data["Average"])
plt.title("Distribution of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
data.sort_values(by="Average", ascending=False).head()
```
